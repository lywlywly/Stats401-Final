```python
import pandas as pd
import os
```

```python
dir = "./Data01/"
csv_files = [file for file in os.listdir(dir) if file.endswith('.csv')]
```

```python
num_cnts = []
total_cnts = []

for csv_file in csv_files:
    path = os.path.join(dir, csv_file)
    df = pd.read_csv(path)
    #df = df.sort_values(by='gacha_type')
    #df = df.reset_index(drop=True)

    num_cnt = 0
    count = 1
    counts = []
    for rank_type in df['rank_type']:
        if rank_type == 5:
            num_cnt = num_cnt + 1
            counts.append(count)
            num_cnts.append(num_cnt)
            total_cnts.append(count)
            count = 1
        else:
            counts.append(count)
            count += 1
    df["count"] = counts
```

```python
df_sort = df.sort_values(by=['gacha_type', 'gacha_time'])
df_sort = df_sort.reset_index(drop=True)
#df_sort.head(60)

count = 1
count_list = []
temp = df_sort.iloc[0, 1]

for i in range(df_sort.shape[0]):

    if df_sort.iloc[i, 3] == 5:
        count_list.append(count)
        count = 1
    
    elif df_sort.iloc[i, 1] != temp:
        count = 1
        count_list.append(count)
        temp = df_sort.iloc[i, 1]
    
    else:
        count_list.append(count)
        count += 1

df_sort["counts"] = count_list
```

```python
df = pd.DataFrame({"num_cnt": num_cnts, "draw_cnt": total_cnts})
```
