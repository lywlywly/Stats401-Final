```python
import numpy as np
import pandas as pd

import os
```

```python
# 收集整理数据
# 初始化一个空的DataFrame
all_data_list = []

# 循环遍历每个文件夹
for folder_num in range(1, 176):
    folder_name = str(folder_num).zfill(4)
    for file_name in ["gacha100", "gacha200", "gacha301", "gacha302"]:
        
        file_path = os.path.join("Data03", folder_name, file_name + ".csv")
        print(file_path)
        # 检查文件是否存在
        if os.path.exists(file_path) and os.path.getsize(file_path) > 0:
            
            df = pd.read_csv(file_path)
            
            #if df.empty: continue

            # 添加"Pool"列
            pool_value = int(file_name.replace("gacha", ""))
            df["Pool"] = pool_value
            
            all_data_list.append(df)

all_data = pd.concat(all_data_list, ignore_index=True)
all_data.to_csv("all_data.csv", index=False)
```

```python
all_data = pd.read_csv("all_data.csv")
```

```python
# 处理数据
# 第一步：改变列名
all_data.columns = ['Time', 'Name', 'Type', 'Star', 'Pool']

# 第二步：处理时间
all_data['Time'] = pd.to_datetime(all_data['Time']).dt.date

# 第三步：处理抽卡类型
all_data['Type'] = all_data['Type'].replace({'武器': 1, '角色': 2})
```
